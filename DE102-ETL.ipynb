```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#extract menjadi data frame
df = pd.read_csv('Salary Prediction.csv')
df.head(1)
```

```python
#create engine connection to postgres
engine = create_engine('postgresql://postgres:postgres@localhost:5432/postgres')
```

```python
#Load to database postgres sql
df.to_sql('salary_prediction',con=engine, if_exists='append', index=False)
```
