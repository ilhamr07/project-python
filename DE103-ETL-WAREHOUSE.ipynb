```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#connection
engine = create_engine('postgresql://postgres:postgres@localhost:5432/postgres')
connection = engine.connect()
```

```python
#connection
engine2 = create_engine('postgresql://posgres2:postgres@localhost:5435/posgres')
connection = engine.connect()
```

```python
#query
my_query = 'SELECT * FROM salary_prediction'
```

```python
#transform data menjadi data frame
df = pd.read_sql_query(my_query,con=connection)
```

```python
df = df.dropna(how='any',axis=0) 

# df.head()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['DOJ'] = pd.to_datetime(df['DOJ']).dt.strftime('%Y-%m-%d %H:%M:%S')
df['CURRENT DATE'] = pd.to_datetime(df['CURRENT DATE']).dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
df.head()
```

```python
df.to_sql('salary_prediction_warehouse',con=engine2, if_exists='append', index=False)
```
